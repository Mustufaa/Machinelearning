# DBSCAN - Density based spatial clustering of applications with noise

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
df= pd.read_csv('Mall_Customers.csv')
X = df.iloc[:, [3, 4]].values
```

```python
X = StandardScaler().fit_transform(X)
```

- eps = neighborhood radius : How close two  point must be in the cluster
- min_samples = Minmum points in the cluster : The minimum number of points needed to form a cluster

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels = dbscan.fit_predict(X)
```

```python
# Visualising the clusters
plt.scatter(X[:,0], X[:,1], c=lables, cmap='viridis',s=50)
plt.title('DBSCAN Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
#Finding outliers (Label -1 means outlier)
outliers = X[labels == -1]
print(f"Total Outliers Detected: {len(outliers)}")
```

```python
plt.scatter(X[:,0], X[:,1], c=labels, cmap='viridis',s=50, label="Clusters")
plt.scatter(outliers[:,0], outliers[:,1], c='red',s=100, label="Outliers", marker='x')
plt.title('DBSCAN Clustering of Customer')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```
